```python
import pandas as pd

def unique_birds(filename):
    data = pd.read_csv(filename)
    
    import pdb; pdb.set_trace()

    unique_species = len(set(data['taxon_species_name']))
    unique_genera = len(set(data['taxon_genus_name']))
    unique_families = len(set(data['taxon_family_name']))
    unique_orders = len(set(data['taxon_order_name']))
    
    return '''In the LA River Watershed, according to the dataset:
              There are {} unique species of birds.
              There are {} unique genera of birds.
              There are {} unique families of birds.
              And, there are {} unique orders of birds.'''.format(unique_species, unique_genera, unique_families, unique_orders)

unique_birds('../data/iNaturalist_birds_LARW.csv')
```
